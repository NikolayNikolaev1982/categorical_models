## Proprieta distributive:

Ipotessi (d): $\epsilon \sum N_n(0, \sigma^2I_n)$ - $\epsilon_i$-IID per ogni i

-> $Y\sim N_n(X\beta, \sigma^2I_n)$ -> Y_i indipendenti ma non identicamente distribuite

- ogni i=1...n

=> Dsitribuzione di $B^o$

$$B^o_{p*1}= GX^tY \sim N_p^{-}(E[B^o], V[B^o])$$
- dove $E[B^o]= GX^tX\beta$
- $V[B^o]= \sigma^2 GX^tXG^t$

Ma $V[B^o]$ non ha rango pieno

$$rank(V[B^o])=ran(GX^tXG^t)= rank(X)=k < p$$

-> $det(V[B^o])= 0$ [V[B^o]e' singolare]

-> non esiste l'iversa di $V[B^o]$

- non esiste fun.densita normale multivariata
($Np^-$: fun.densita normale degenere)

Tutavia l'inversa e' rivolta al BLUE r-dimensionale $QB^o$. Sotto l'ipotesi (d):

$$Q_{r*1}B^o=QGX^tY\sim N_r (Q\beta, \sigma^2QGQ^t)$$

dove: $1\le r\le k$ e $rank[V[QB^o]]=r$



## Stimatore per l'ignota varianza $\sigma^2$

- punto di partenza: v.c. Devianza Residua

$$SSE=\sum_{i=1}^n (Y_i-\hat{Y_i})^2= Y^tNY$$

[N=I_n-XGX^t]

-> poiche le v.c. SSE e' una forma quadratica di un vettore casuale normale, si applicano direttamente i teoremi dche riguardano la legge di distribuzioni di tali vettori

$E(SSE)= E[Y^tNY]= [n-rank(X)]\sigma^2= (n-k)\sigma^2$
- ogni $\sigma^2>0$

Stimatore: varianza campionaria corretta

$$f^2= \sum_{i=1}^n (Y_i-\hat{Y_i})^2/ (n-k)= SSE/(n-k)= \frac{T^tNY}{n-k}$$

-> $E[f^2]= \sigma^2$
  - ogni $\sigma^2>0$

Osservazione: Nel CLM

$f^2= \frac{SSE}{n-(k+1)}= \frac{SSE}{n-rank(X)}$
- dove rank(X)= k+1=p
- X matrice (n*p) a rango pieno



## Legge di distribuzione dello stimatore $f^2$

- Risultato di riferente:

$$\frac{SSE}{\sigma^2}\sim \chi_{n-rank(X)}^2$$

dove:
- $\chi_{n-rank(X)}^2$ e' la v.c. CHi-Quadrato con n-rank(X) degree of freedom

$df(SSE)= n-rank(X)$
-> $SSE \sim \sigma^2 \chi_{n-rank(X)}^2$
-> $f^2= \frac{SSE}{n-k}\sim \frac{\sigma^2}{n-k}\chi_{n-k}^2$

## Intervallo di confidenza per $q^t\beta$
- metodo del Quantita Pivotali

Si parte da una legge di disribuzione di riferimento:

$q^tB^o \sim N(q^t\beta; \sigma^2q^t Gq)$
-> $\frac{q^tB^o-q^t\beta}{\sqrt{\sigma^2q^tGq}}\sim N(0,1)$

dove:
- $\sqrt{q^tB^o}= \sqrt{\sigma^2q^tGq}= SE[q^tB^o]$ standard error

-> Problema: $\sigma^2$ e' ignoto:

Si considera al suo posto lo stimatore $f^2$
- si ottine la seguente leggei di distribuzione:

$$\frac{q^tB^o-q^t\beta}{\sqrt{f^2q^tGq}}\sim T_{n-rank(X)}$$

- $T_{n-rank(X)}$ e' la v.c. t di student con df(SSE)= n-rank(X)


Costruzione dello stimatore intervallare per $q^t\beta$
- si fissa il livello di confidenza:
(1-\alpha)100%  con 0<\alpha<1

di soluto: \alpha=0.05-> (1-0.05)100%=95%

$$P_r[-t_{\frac{\alpha}{2}} \le T_{n-k}\le t_{\frac{\alpha}{2}}]= 1-\alpha$$
dove:
- $Pr[T_{n-k} \ge t_{\frac{\alpha}{2}}]= Pr[T_{n-k}\le -t_{\frac{\alpha}{2}}]= \frac{\alpha}{2}$

Sostituendo a $T_{n-k}$ il rapporto definito in precedenza si ottiene:

$$Pr[-t_{\frac{\alpha}{2}}\le \frac{q^tB^o-q^t\beta}{\sqrt{f^2q^tGq}}\le -t_{\frac{\alpha}{2}}]= 1-\alpha$$

$$Pr[-t_{\frac{\alpha}{2}}\sqrt{f^2q^tGq}\le q^tB^o-q^t\beta \le t_{\frac{\alpha}{2}}]=1-\alpha$$

$$Pr[-q^tB^o-t_{\frac{\alpha}{2}}\sqrt{f^2q^tGq}\le -q^t\beta \le -q^tB^o+t_{\frac{\alpha}{2}}\sqrt{f^2q^tGq}]=1-\alpha$$

$$Pr[q^tB^o-t_{\frac{\alpha}{2}}\sqrt{f^2q^tGq}\le -q^t\beta \le q^tB^o+t_{\frac{\alpha}{2}}\sqrt{f^2q^tGq}]=1-\alpha$$

- stimatore intervallare o intervallo casuale per $q^t\beta$

$$[q^tB^o-t_{\frac{\alpha}{2}}\sqrt{f^2q^tGq} ; q^tB^o+t_{\frac{\alpha}{2}}\sqrt{f^2q^tGq}]$$
- contiene con probabilita $1-\alpha$ l'ignota f.stim $q^t\beta$

- dallo stimatore intervallare si ottiene l'intervallo di confidenza sul campione al livello di confidenza $1-\alpha$:

$$[q^tB^o-t_{\frac{\alpha}{2}}\sqrt{s^2q^tGq}, q^tB^o+t_{\frac{\alpha}{2}}\sqrt{s^2q^tGq}]$$

oppure:

$q^t\beta^o \pm \sqrt{s^2q^tGq}$
- dove: $SE(q^t\beta^o)=\sqrt{s^2q^tGq}$ - standard error stimato sul campione

caso particolare modello 1-way ANOVA

$$SST=SSE+SSR$$

Specificazione della statistica test $F_{r, n-rank(X)}$ al caso in esame
- leggi di distribuzione sotto l'ipotesi (d)
- v.c. Devianza Residua

$$\frac{SSE}{\sigma^2}= \frac{\sum_{i=1}^n (Y_i - \hat{Y_i})^2}{\sigma^2} \sim \chi_{n-rank(X)}^2$$

-v.c. Devianza Spiegata

$$\frac{SSR}{\sigma^2}= \frac{\sum_{i=1}^n (\hat{Y_i} - \bar{Y_i})^2}{\sigma^2} \sim \chi_{n-rank(X)-1}^2$$
- sotto ipotesi: $H_0:L\beta$

- v.c. Devianza Totale:

$$\frac{SST}{\sigma^2}= \frac{\sum_{i=1}^n (Y_i - \bar{Y_i})^2}{\sigma^2} \sim \chi_{n-1}^2$$
- sotto ipotesi: $H_0:L\beta$

Varianze:
- varianza residua: $MSE=\frac{SSE}{n-rank(X)}=f^2$
- spiegata: $MSR= \frac{SSR}{rank(X)-1}$
- totale: $MST=\frac{SST}{n-1}$

|Fonte di variabilita|Devianze|df|Varianze|F_e|

|Modello |SSR       |rank(X)-1|MSR|\frac{MSR}{MSE}|
|Errore  |SSE       |n-rank(X)|MSE|p-value|
|Totale(correto)|SST|n-1|      MST||

$F_e=\frac{MSR}{MSE}$ valore empirico della statistica test $F_{k-1;n-k}$

- r=k-1 opotesi LIN

$$\frac{\frac{\chi_{k-1}^2}{k-1}  }{\frac{\chi_{n-k}^2}{n-k}  }=\frac{\frac{SSR}{k-1}}{\frac{SSE}{n-k}} \sim F_{k-1;n-k}$$
- sotto $H_0:L\beta=0$
- $\frac{\frac{SSR}{k-1}}{\frac{SSE}{n-k}}= \frac{MSR}{MSE}$
- sul campione si ha il valore empirico F_e cha ha associato il p-value

p-value: $Pr[F_{k-1;n-k} \ge F_e]$

La v.d.i. viene condotta confrontando il p-value con il livello di significativita nominale $\alpha$ (di solito $\alpha=0.05$) tale che:

$$Pr[F_{k-1;n-k} \ge F_e]= \alpha$$

Quindi se:
- $p-val > \alpha$ -> si accetta $H_0$
- $p-val \le \alpha$ -> si rifiuta $H_0$
