## Significato dell'ipotesi sull'interazione

- Punto di partenza: si considerano esclusivamente SItuazini AMMISSIBILI interpretate dei MODELI GERARCHICI

Def: informale di Mod/Gerarchico: data p var.categ il modello e' detto Gerarchico se la presenza di un parametro di interazione di ordine $h$ ($h\ge 0$) relativo ad un sottoinsieme delle p var.implica la presenza di tutti i parametri di interazione di ordine inferiore ad h con riferimento a quello stesso sottoinsieme di variabili.

- 2-way ANOVA:
  - $\alpha_i, \beta_j$ effetti principali o interazioni di ordine 0 (h=0)
  - $(\alpha\beta)_{ij}$ interazione dippia o interazione di ordina 1  (h=1)

-> Il modello ANOVA a due vie e' gerarchico <->  la presenza dell'interazione $(\alpha \beta)_{ij}$ ogni i,j implica la presenza di tutti gli effetti $\alpha_i$ e $\beta_j$ ogni i, ogni j

$$\{A, B, AB \}\text{-> modello ANOVA a due vie completo}$$


Altrimenti il modelo e' detto NON gerarchico:

$\{AB\}- Y_{ijl}= \mu+(\alpha\beta)_{ij} + \epsilon_{ijl}$
- ogni i,j,l

Oppure:

$$\{A, AB\}-> Y_{ijl}= \mu +\alpha_i +(\alpha\beta)_{ij} + \epsilon_{ijl}$$ ogni i,j,l

1. Somme dei quadrati di Tipo- I - SS1 - Type I sum of Squars:

|Fonte di Variabilita|SS1|df|Reduction Notation- diferenza fra riduzione nelle SS||
|---|---|---|---|---|
|A|SS1A|a-1|$R(\alpha|\mu)= R(\mu, \alpha)- R(\mu)$||
|B|SS1B|b-1|$R(\beta|\mu)= R(\mu,\alpha,  \beta)- R(\mu, \alpha)$|
|AB|SS1AB|(a-1)(b-1)|$R(\alpha \beta|\mu, \alpha, \beta)= R(\mu, \alpha, \beta, \alpha, \beta)- R(\mu, \alpha, \beta)$|

Carateristiche delle SS1
- sono dette some sequenziali
- si basano implicitamente su una cosatruzione  "in avanti" del modello
- producono una scomposizione additiva di SSR
- dipendono dall'ordine di inserimento degli effetti nel modello (tranne nel caso di dati bilanciati)
- dipendono dai conteggi di cella, ossia son in funzione delle numerosita $n_{ij}$ delle combinazioni $(A_i, B_j)$ per i=1...a, j=1...b
- si esprimono analiticamente come diferenza fra riduzione nelle SS


-> SS1: inferenza relativa ad una popolazione non bilanciata che ha la stessa struttura del campione

Statistica test- F di Fisher

$$H_0^A-> F_e=\frac{MS1A}{MSE}->F_{a-1; n-ab}$$
- se $H_0^A vera$
- $MS1A= \frac{SS1A}{a-1}$

$$H_0^B-> F_e=\frac{MS1B}{MSE}->F_{b-1; n-ab}$$
- se $H_0^B vera$
- $MS1B= \frac{SS1B}{b-1}$

$$H_0^{AB}-> F_e=\frac{MS1AB}{MSE}->F_{(a-1)(b-1); n-ab}$$
- se $H_0^{AB} vera$
- $MS1AB= \frac{SS1AB}{(a-1)(b-1)}$

Lettura delle v.d.i. nel caso dei dati NON bilanciati:

$$H_0^A: \alpha_1^*-\alpha_a^*= ...=\alpha_{a-1}^*-\alpha_a^*$$

$$H_0^B: \beta_1^*-\beta_b^*= ...=\beta_{b-1}^*-\beta_b^*$$
- data A nel modello

$$H_0^{AB}: (\alpha\beta)_{ij}-(\alpha\beta)_{sq}= (\alpha\beta)_{sj}-(\alpha\beta)_{sq}$$
- ogni $i\ne s, j\ne q$
- date A e B nel modello attraverso effetti principali

## Tipologia di Contrasti:

Riparametrizzazione del modello ANOVA a due vie complleto:

$$Y_{ijl}=\mu+\alpha_i +\beta_j +(\alpha\beta)_{ij} + \epsilon_{ijl}$$
- ogni i,j,l
- si po ne $\theta_{ij}= \alpha_i +\beta_j+(\alpha \beta)_{ij}$
- ogni i,j

Quindi:
$$Y_{ijl}= \mu +\theta_{ij}+\epsilon_{ijl}$$-  ogni i,j,l

Dove:
- $\theta_{ij}$ effetto della combinazione di categorie ($A_i, B_j$) anche detta "tratamento i,j" per semplificare il linguaggio
- $\theta_{ij}$= effetto del trattamento i,j ogni i,j

Oss: il modello riparamerizzato si puo anche intendere quele modello ANOVA ad 1 via che riguarda la variabile AB con categorie date dai trattamenti

11,12,13...21,22,23....ij...ab

$$\mu_{ij}= \mu + \alpha_i+\beta_j +(\alpha\beta)_{ij}=\mu + \theta_{ij}$$

<=> $\theta_{ij}= \mu_{ij}-\mu$ ogni i,j

e' sempre possibile ricavare $\mu_{ij}$da $\theta_{ij}$ e viceversa

Tutte le tipologie di contrati rientrano nella seguente forma generale:

$$L=\sum_{i=1}^a\sum_{j=1}^b d_{ij}\theta_{ij}$$

con:
- $\sum_{i=1}^a d_{ij}=0$ ogni j
- $\sum_{i=1}^a \sum_{j=1}^b d_{ij}=0$

Osservazione: L e' certamente una f.tim Basta porre: $y_{ij.}=d_{ij}$ nella forma generale delle f.stim dell ANOVA a due vie

## Tipologia principali di contrasti
- che rientrano nella definizione generale

1. Contrasti dei trattamenti:

Es: $L=\theta_{ij}-\theta_{sh}$
$$=\alpha_{i}+\beta_j + (\alpha \beta)_{ij} - \alpha_{s}+\beta_h + (\alpha \beta)_{sh}= (\alpha_i-\alpha_s)+(\beta_j - \beta_h)+[(\alpha\beta)_{ij}-(\alpha\beta)_{sh}]$$


Oss: i contrasti dei trattamenti si esprimono anche come contrasti delle medie congiunte. Nel caso del'esempio:

$L=\mu_{ij}-\mu_{sh}=\mu +\theta_{ij}-\mu -\theta_{sh}= \theta_{ij}-\theta_{sh}$

in gemeral;e:

$$L=\sum_i \sum_j d_{ij}\mu_{ij}= \sum_i \sum_j d_{ij} (\mu+\theta_{ij})= \mu \sum_i \sum_j d_{ij}+\sum_i \sum_j d_{ij} \theta_{ij}= $$

$$\sum_i \sum_j d_{ij} \theta_{ij}$$


2. Contrasti dell'interazione:

In generale: $L=\sum_i \sum_j d_{ig}(\alpha\beta)_{ij}$

Es: $L=[(\alpha \beta)_{ij}-(\alpha\beta)_{ij}]- [(\alpha \beta)_{iq}-(\alpha\beta)_{sq}]$

Oss: L rispetta la def. generale di contrasto. Infatti:

$$\sum_{h=1}^a d_{hj}= 0+0+...+1+...+(-1)+...+0=0$$
- per j fisato
- h=i=1
- h=s=(-1)

In modo analogo:

$\sum_{h=1}^a d_{hq}= 0+0+...+(-1)+...1+..0=0$
- per q fissato
- h=i=-1
- h=s=1

Inoltre:

$\sum_{m=1}^b d_{im}= 0+0...+1+...+(-1)+..0=0$
- per i fissato
- m=j=1
- m=q=-1

In modo analogo:
$\sum_{m=1}^b d_{sm}= 0+0+...(-1)+...1+...+0=0$
- per s fisato
- m=j=-1
- m=q=1

In tutti gli alri casi: $d_{hm}= 0$ ogni $(h,m)\ne (i,j)(i,q)(s,j)(s,q)$

Contro esempio:

$$L=(\alpha\beta)_{ij}-2(\alpha\beta)_{iq}+(\alpha\beta)_{sq}$$

Non e' un contrasto e non e' stimabile

Infatti, per valendo:

$\sum_{h=1}^a \sum_{m=1}^b d_{hm}=0+..+1+...+(-2)+...+1+...+0=0$

si ha:
- $\sum_{h=1}^a d_{hj}= 0+..+1+...+0=1 \ne 0$
- h=i=1
- per j fissato

$\sum_{h=1}^a d_{hq}= 0+...+(-1)+...+1+...+0=-1 \ne 0$
- h=i=-2
- h=s=1
- per q fissato


Osservazione: $L=[(\alpha\beta)_{ij}- (\alpha\beta)]_{sj}]- [(\alpha\beta)_{iq}- (\alpha\beta)]_{sq}]$

puo essere scritta anche come:
- $L=[\theta_{ij}-\theta_{sj}]- [\theta_{iq}-\theta_{sq}]= [\mu_{ij}-\mu_{sj}]-[\mu_{iq}-\mu_{sq}]$


Esempioo: siano A e B due var.categ con a=3, b=2 categorie. Si considera:

$L=(\mu_{11}-\mu_{21})-(\mu_{12}-\mu_{22})$
