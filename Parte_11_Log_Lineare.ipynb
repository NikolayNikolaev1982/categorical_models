$$\theta_i = ln(\mu_i)= \eta_i=x_i^t\beta$$

Modello Logistioco Binomiale

Punto di partenza: variabile dipendente dicotomica S con due categorie

- $s$ evento sucesso - $P[S=s]=\pi$
- $\bar{s}$ evento insucesso -> $P[S=\bar{s}]= 1-\pi$
- $0<\pi<1$

Si ricorre alla v.c. Y:

y=1: se S=2 $P[Y=1]=\pi$

y=0 se $S=\bar{s}$ -$P[Y=0]=1-\pi$


-> $Y\sim Berouli(\pi)$

$$f(y, \pi)= \pi^y(1-\pi)^{1-y}$$
- y=0,1
- $0<\pi<1$

$$Y\sim Ber(\pi) \in FE$$ con

$\theta= ln(\frac{\pi}{1-\pi})logit(\pi)$
- $\theta \in R$

- $b(\theta= ln(1+e^{\theta}))$
- $a(\phi)= 1$
- $c(y, \phi)= 0$

inparticolare:

$\theta= ln(\frac{\pi}{1-\pi})-> \frac{\pi}{1-\pi}=e^{\theta}= odds(\pi)$

- $\pi=P[Y=1]=\frac{e^{\theta}}{1+e^{\theta}}$

- $1-\pi= P[Y=0]= \frac{1}{1+e^{\theta}}$

Valore attesso e varianza di Y:

$$E[Y]= \frac{db(\theta)}{d\theta} =\frac{d}{d\theta}[ln(1+e^{\theta})]= \frac{e^{\theta}}{1+e^{\theta}}= \pi=P[Y=1]=\mu$$


$$Var[Y]= a(\phi)V(\mu)= V(\mu)= \frac{d^2b(\theta)}{d\theta^2}= \frac{d}{d\theta}[\frac{e^{\theta}}{1+^{\theta}}]= \frac{e^{\theta}}{1+e^{\theta}} \frac{1}{1+e^{\theta}}= $$
- $a(\phi)=1$

$$= \pi(1-\pi)$$

Generalizzaione a $n$ prove indipendenti
->

$$Y\sim Binom(n,\pi)$$

$$f(y, n, \pi)= \binom{n}{y}\pi^y (1-\pi)^{n-y}$$
- y=0,1..n
- 0<\pi<1

Osservazione: n=1

$Ber(\pi)= Binom(1,\pi)$

$Y \sim Bin(n, \pi)\in FE$ con

$\theta= ln(\frac{\pi}{1-\pi})=logiy(\pi)$

$b(\theta)= n.ln(1+e^{\theta})$

$a(\phi)=1$

$c(y, \phi)= ln\binom{n}{y}$
- dove $\binom{n}{y}= $\frac{n!}{y!(n-y)!}

Valore atteso e varianza:

$$E[Y]=\frac{db(\theta)}{d\theta} =n\frac{e^{\theta}}{1+e^{\theta}}= n\pi(=\mu)$$


$$Var[Y]= V(\mu)= n\frac{e^{\theta}}{1+e^{\theta}}\frac{1}{1+e^{\theta}}= n\pi(1-\pi)$$
