Prrsupposti per ;'applicazione del model ANCOVA:
1. modello "equal-slop" , ossia i coefficienti di regresione categoria si intendono (approssitivamente) uguali fra loro

2. relazione  lineare tendenzialmente intesa fra Y e Z

Equazione del modello ANCOVA ad una via e con una covariata
- due formulazioni del modello:

1. formulazione "non tradizionale"

$$Y_{ij}= \mu_j + bz_{ij}+\epsilon_{ij}$$

- $bz_{ij}$

$$= \mu+\alpha_j +bz_{ij}+\epsilon_{ij}$$
- ogni i=1..n, j=1...k

dove:
$$\mu_j=\mu+\alpha_j<-> \alpha_j=\mu_j-\mu$$
- ogni j

- b: coeff.di regressione di Y su Z comune a tutte le categorie $
A_j$ di A

2. Fromulazione "tradizionale":

$$Y_{ij}= \mu^*+\alpha_j + b(z_{ij}-\bar{z})+\epsilon_{ij}$$
- ogni, i, ogni j

dove:

$\bar{z}= \frac{1}{n}\sum_{j=1}^k \sum_{i=1}^{n_{j}}z_{ij}= \bar{z_{..}}$
- ..: intende la media complesssiva

$\mu^*= \mu +b\bar{z}$ <-> $\mu=\mu^*-b\bar{z}$
- costannte del modello confrontando le due formulazioni


Verifica dell'equazione fra le formulazioni
- si sostituisce nella (2) a $\mu^*$ la sua relazione con $\mu$:

$$Y_{ij}= (\mu +b\bar{z})+\alpha_j + bz_{ij}-b\bar{z}+\epsilon_{ij}=$$

$$= \mu+\alpha_j +bz_{ij}+\epsilon_{ij}$$
-> si e' ottenuta la formulazione (1)

Oss: significato di $\mu_j$
- nella formulazione (1)
- $E[\epsilon_{ij}]=0$ ogni i, ogni j
-> $E[Y_{ij|z_{ij}}]= \mu+\alpha_j +bz_{ij}=\mu_j + bz_{ij}$ - ogni i, ogni j

- retta di regressione di Y su Z entro ciascuna categoria $A_j$ di A

-> $E[Y_{ij}|z_{ij}=0]=\mu_{j}$ ogni i, ogni j

-> $\mu_j$ e' la media di Y entro la categoria $A_j$ quando Z assume valore nullo

Nella formulazione (2)

$E[\epsilon_{ij}]=0$ ogni  i, ogni j
- $E[Y_{ij}|z_{ij}]= \mu^* +\alpha_j +b(z_{ij}-\bar{z})$ ogni i, ogni j
- retta di regresione di Y sulla variabile Z centrata, ossia: $Z-\bar{z}$ entro ciascuna categoria $A_j$ di A

-> $$EY_{ij}|z_{ij}=\bar{z}]=\mu^* +\alpha_j +b.0=\mu^* +\alpha_j=$$

$$=\mu+b\bar{z}+\alpha_j=\mu+\alpha_j+b\bar{z}= $$

$$\mu_j+b\bar{z}$$
- ogni i, ogni j

-> Posto : $z_{ij}=\bar{z}=0$ ogni i, j

$$E[Y_{IJ}|\bar{z}=0]=\mu_j$$ ogni i, j

- $\mu_j$ e' la media di Y entro la categoria $A_j$ in corrispondenza di valori tutti nulli per Z (in modo che anche la media complessiva di Z sia nulla)





## Modello ANCOVA in forma matricale


equaione:$$Y_{n*1}= X_{n*p}\beta_{p*1}+Z_{n*c}b_{c*1}+\epsilon_{n*1}$$

X: matrice disegno relativa alla variabili categoriali
- contiene la colonna unitaria che corrisponde al termine costante

Z- matrice disegno relative alle c covariate
- non contiene la colonna unitaria

Osservazione: l'equazione del modello ANCOVA rientra nella formulazione del GLM

Infatti , indicati con:

$$W=[X;A]\text{ a } [n*(p*c)$$ dimensioni

$η=[\beta; b]^t \text{ a }[(p*c)*1]$ dimenioni


-> $Y=X\beta+Zb+\epsilon= [X;Z][\beta;b]^t+\epsilon=$

$=W\eta+\epsilon$

Ipotesi del modello ANCOVA:
- a: $E[\epsilon]=0$
- b: $V[\epsilon]=\sigma^2I_n$- comprende due ..
- c: X e Z matrici deterministiche

  - rank(X)=k<p - rango non pieno
  - rank(Z)= c - rango pieno
  - rank(X)+rank(Z)= k+c <p+c<n

- d: necessaria per inferenza, ma non per la specificazione del modello. $\epsilon\sim N(0, \sigma^2I_n)$

### (V) medie corrette per la presenza delle covariate - Adjusted Mens

ANCOVA: $Y_{ij}+\alpha_j + bz_{ij}+\epsilon_{ij}$ ogni i,j

ANOVA: $Y_{ij}+\alpha_j+\epsilon_{ij}$ ogni i,j

Modello ANOOVA a 1-way : Stima corretta pr $\mu_j$

$$\bar{y_{.j}}=\frac{1}{n_j}\sum_{i=1}^{n_j} y_{ij}->$$

$$->\bar{Y_{.j}}=\frac{1}{n_j} \sum_{i=1}^{n_j} Y_{ij}$$


$$E[\bar{Y_{.j}}]= E[\frac{1}{n_j} \sum_{i=1}^{n_j} Y_{ij}]= \frac{1}{n_j}E[\sum_{i=1}^{n_j} Y_{ij}]= $$
- $E[Y_{ij}=\mu+\alpha_j]$ ogni i,j

$$\frac{1}{n_j} \sum_{i=1}^{n_j} (\mu+\alpha_j)= \frac{1}{n_j}(\mu+\alpha_j)= \mu+\alpha_j=\mu_j$$ ogni $\mu_j$

In SAS, nel modello ANOVA ad 1 via

Statement MEANS => $\bar{y_{.j}} ≡$ statmens LSMEANS-> $\hat{\mu_j}$


Nel modelo ANCOVA $\bar{Y_{.j}}$ NON e' stimatore corretto per $\mu_j$. infatti:

$$E[\bar{Y_{.j}}]= \frac{1}{n_j}\sum_{i=1}^{n_j}E[Y_{ij}]= \frac{1}{n_j} (\mu +\alpha_j + bz_{ij})= $$
- $E[Y_{ij}]= \mu +\alpha_j +bz_{ij}$ ogn i,j

$$= \frac{1}{n_j} n_j\mu + \frac{1}{n_j}n_j \alpha_j + \frac{b}{n_j}\sum_{i=1}^{n_j}= z_{ij}= $$

Osserazione: $\frac{1}{n_j}\sum_i z_{ij}=   \bar{z_{.j}} $ media di Z entro $A_j$ di A , ogni j

$$= \mu +\alpha_j +b\bar{z_{.j}}= \mu_j +b \bar{z_{.j}}\ne \mu_j$$


Risultato fornito da SAS con lo statement MENAS -> Unadjusted Means : $\bar{y_{.j}}$


Adjusted means: Statement LSMEANS

$\hat{\mu_j}=         \bar{Y_{.j}}- \hat{B}\bar{z_{.j}}$ ogni j

$$E[\hat{\mu_j}]= E[\bar{Y_{.j}}- \hat{B}\bar{z_{.j}}]= $$

$$E[\bar{Y_{.j}}]- E[\hat{B}\bar{z_{.j}}]= $$

$$= \mu_j+b\bar{z_{.j}}-b\bar{z_{.j}}=\mu_j$$

$\hat{\mu_j}$ e' coretto per $\mu_j$
